Cul es la distribucin de actividad fsica semanal?

Las personas con ms actividad fsica tienen menor riesgo de diabetes?

```python
sns.histplot(
    data=data,
    x="physical_activity_minutes_per_week",
    hue="diagnosed_diabetes",  # separa por esta columna
    bins=20,
    palette=["skyblue", "orange"],  
    alpha=0.6, 
    multiple="stack"  # puede ser "stack" o "dodge"
)

plt.title('Actividad fsica por diagnstico de diabetes')
plt.xlabel('Minutos de actividad fsica por semana')
plt.ylabel('Cantidad de personas')
plt.show()
```

```python
data.groupby('diagnosed_diabetes')[["physical_activity_minutes_per_week",'diabetes_risk_score']].mean().reset_index()
```

```python
from scipy.stats import ttest_ind

# Separamos en grupos
grupo_no_diabetes = data[data['diagnosed_diabetes'] == 0]['physical_activity_minutes_per_week']
grupo_diabetes = data[data['diagnosed_diabetes'] == 1]['physical_activity_minutes_per_week']

# T-test independiente
t_stat, p_value = ttest_ind(grupo_no_diabetes, grupo_diabetes, equal_var=False)  # Welch's t-test

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

El anlisis muestra que las personas que no tienen diabetes realizan en promedio ms actividad fsica semanal (~ 129 min) que las personas con diabetes (~112 min). El t-test confirm que esta diferencia es altamente significativa (t  31.27, p  2.510), lo que indica que no es fruto del azar.

Por lo tanto, se puede afirmar que, en este conjunto de datos, mayor actividad fsica se asocia con menor riesgo de diabetes. Este resultado coincide con la evidencia clnica conocida de que la actividad fsica protege contra la aparicin de diabetes tipo 2.

```python
import statsmodels.api as sm

X = data[['physical_activity_minutes_per_week', 'diet_score', 'sleep_hours_per_day']]
y = data['diabetes_risk_score']

X = sm.add_constant(X)  # agrega intercepto
model = sm.OLS(y, X).fit()
print(model.summary())
```

La regresin lineal mltiple muestra que la actividad fsica y la dieta son los factores ms importantes para reducir el riesgo de diabetes. Cada minuto extra de actividad fsica disminuye el diabetes_risk_score en 0.037 unidades y mejorar la dieta reduce el riesgo en 0.74 unidades, ambos resultados estadsticamente significativos. En cambio, las horas de sueo no tienen un efecto detectable sobre el riesgo (p=0.569). Aunque el modelo es significativo en conjunto, solo explica alrededor del 14% de la variabilidad del riesgo (R=0.142), lo que indica que hay otros factores importantes no incluidos en este anlisis.
