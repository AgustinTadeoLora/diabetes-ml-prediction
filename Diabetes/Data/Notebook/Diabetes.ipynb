¿Cuál es la distribución de actividad física semanal?

¿Las personas con más actividad física tienen menor riesgo de diabetes?

```python
sns.histplot(
    data=data,
    x="physical_activity_minutes_per_week",
    hue="diagnosed_diabetes",  # separa por esta columna
    bins=20,
    palette=["skyblue", "orange"],  
    alpha=0.6, 
    multiple="stack"  # puede ser "stack" o "dodge"
)

plt.title('Actividad física por diagnóstico de diabetes')
plt.xlabel('Minutos de actividad física por semana')
plt.ylabel('Cantidad de personas')
plt.show()
```

```python
data.groupby('diagnosed_diabetes')[["physical_activity_minutes_per_week",'diabetes_risk_score']].mean().reset_index()
```

```python
from scipy.stats import ttest_ind

# Separamos en grupos
grupo_no_diabetes = data[data['diagnosed_diabetes'] == 0]['physical_activity_minutes_per_week']
grupo_diabetes = data[data['diagnosed_diabetes'] == 1]['physical_activity_minutes_per_week']

# T-test independiente
t_stat, p_value = ttest_ind(grupo_no_diabetes, grupo_diabetes, equal_var=False)  # Welch's t-test

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

El análisis muestra que las personas que no tienen diabetes realizan en promedio más actividad física semanal (~ 129 min) que las personas con diabetes (~112 min). El t-test confirmó que esta diferencia es altamente significativa (t ≈ 31.27, p ≈ 2.5×10⁻²¹³), lo que indica que no es fruto del azar.

Por lo tanto, se puede afirmar que, en este conjunto de datos, mayor actividad física se asocia con menor riesgo de diabetes. Este resultado coincide con la evidencia clínica conocida de que la actividad física protege contra la aparición de diabetes tipo 2.

```python
import statsmodels.api as sm

X = data[['physical_activity_minutes_per_week', 'diet_score', 'sleep_hours_per_day']]
y = data['diabetes_risk_score']

X = sm.add_constant(X)  # agrega intercepto
model = sm.OLS(y, X).fit()
print(model.summary())
```

La regresión lineal múltiple muestra que la actividad física y la dieta son los factores más importantes para reducir el riesgo de diabetes. Cada minuto extra de actividad física disminuye el diabetes_risk_score en 0.037 unidades y mejorar la dieta reduce el riesgo en 0.74 unidades, ambos resultados estadísticamente significativos. En cambio, las horas de sueño no tienen un efecto detectable sobre el riesgo (p=0.569). Aunque el modelo es significativo en conjunto, solo explica alrededor del 14% de la variabilidad del riesgo (R²=0.142), lo que indica que hay otros factores importantes no incluidos en este análisis.

```python
X =  data[[
    'age',
    'alcohol_consumption_per_week',
    'physical_activity_minutes_per_week',
    'diet_score',
    'sleep_hours_per_day',
    'screen_time_hours_per_day',
    'family_history_diabetes',
    'hypertension_history',
    'cardiovascular_history',
    'bmi',
    'waist_to_hip_ratio',
    'systolic_bp',
    'diastolic_bp',
    'heart_rate',
    'cholesterol_total',
    'hdl_cholesterol',
    'ldl_cholesterol',
    'triglycerides',
    'glucose_fasting',
    'glucose_postprandial',
    'insulin_level',
    'hba1c',
    'diabetes_risk_score'
]]
y = data['diagnosed_diabetes']

X = sm.add_constant(X)

model = sm.Logit(y,X)

result = model.fit()

print(result.summary())

odds_ratios = pd.DataFrame({
    'OR': result.params.apply(lambda x: round(np.exp(x), 2)),
    'p-value': result.pvalues
})

odds_ratios = odds_ratios.sort_values(by='OR', ascending=False)

print("\nOdds Ratios y p-values:")
print(odds_ratios)
```

Al construir un modelo de regresión logística incluyendo todas las variables disponibles:

Variables clave:

hba1c y glucose_fasting se destacan claramente como predictores fuertes y significativos del diagnóstico de diabetes (OR ≈ 30.75 y 1.02 respectivamente, p ≈ 0).

Estas variables reflejan directamente el control de glucosa y son las más confiables para identificar riesgo.

Otras variables:

La mayoría de las demás variables, como actividad física, dieta, sueño, colesterol, presión arterial, BMI, antecedentes familiares y cardiovasculares, tienen OR cercanos a 1 y p > 0.05, indicando que no aportan información significativa en presencia de los biomarcadores de glucosa.

Variables con OR > 1 pero p > 0.05 (por ejemplo waist_to_hip_ratio o family_history_diabetes) no se pueden considerar confiables para la predicción.

Aunque muchas variables podrían asociarse con diabetes de forma individual, al analizar todas juntas, solo algunas dominan la predicción.
